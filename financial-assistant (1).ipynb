```python
import pandas as pd

# List 1: Things you spent money on
spending_data = pd.DataFrame({
    'Date': ['2025-07-01', '2025-07-03', '2025-07-10', '2025-07-15', '2025-07-21'],
    'Category': ['Groceries', 'Utilities', 'Streaming', 'Restaurant', 'Internet'],
    'Amount': [3500, 1200, 599, 1100, 799]
})

# List 2: Things you always pay for, like Netflix
subscriptions = pd.DataFrame({
    'Service': ['Netflix', 'Spotify', 'Internet'],
    'Monthly Cost': [599, 199, 799]
})

print("Spending Data:\n", spending_data)
print("\nSubscriptions:\n", subscriptions)
```

```python
category_totals = spending_data.groupby('Category')['Amount'].sum()
print("Total Spent per Category:\n", category_totals)

high_spending = category_totals[category_totals > 2000]
if not high_spending.empty:
    for category, amt in high_spending.items():
        print(f"Recommendation: Try to reduce spending on {category}. You spent ₹{amt} this month!\n")
else:
    print("Spending is balanced. Good job!")
```

```python
negotiable = ['Internet', 'Utilities']

for index, row in spending_data.iterrows():
    if row['Category'] in negotiable and row['Amount'] > 500:
        print(f"Alert: Consider negotiating your {row['Category']} bill of ₹{row['Amount']}. You may get a better deal!")
```

```python
market_rates = {'Netflix': 499, 'Spotify': 149, 'Internet': 699}

for idx, row in subscriptions.iterrows():
    current = row['Monthly Cost']
    market = market_rates.get(row['Service'], current)
    if current > market:
        print(f"Deal found: {row['Service']} can be switched from ₹{current} to ₹{market}/mo. Save ₹{current - market}/mo!")
```

```python
import yfinance as yf

portfolio = pd.DataFrame({
    'Stock': ['AAPL', 'TCS.NS'],
    'Shares': [5, 2]
})

for idx, row in portfolio.iterrows():
    stock = yf.Ticker(row['Stock'])
    price = stock.history(period='1d')['Close'][0]
    value = row['Shares'] * price
    portfolio.loc[idx, 'Current Price'] = price
    portfolio.loc[idx, 'Value'] = value

print("Your Portfolio:\n", portfolio)

total_portfolio = portfolio['Value'].sum()
portfolio['Percent'] = portfolio['Value'] / total_portfolio * 100
heavy = portfolio[portfolio['Percent'] > 70]
if not heavy.empty:
    for idx, row in heavy.iterrows():
        print(f"Rebalance Tip: {row['Stock']} makes up {row['Percent']:.1f}% of your portfolio. Diversify your investments!")
else:
    print("Portfolio is well balanced.")
```
